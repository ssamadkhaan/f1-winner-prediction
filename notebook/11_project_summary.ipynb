# Formula 1 Winner Prediction Project

This project used historical Formula 1 race data to predict race winners using Machine Learning.

The workflow included:

1. Data Collection
2. Feature Engineering
3. Logistic Regression Model
4. Model Evaluation
5. Model Improvement
6. Model Comparison
7. Random Forest Model
8. Feature Importance Analysis
9. 2026 Winner Prediction
10. Hyperparameter Tuning

```python
models = [
    "Logistic Regression",
    "Decision Tree",
    "Random Forest",
    "Tuned Random Forest"
]

accuracy = [
    0.9606,
    0.9946,
    0.9944,
    0.9957
]

for model, score in zip(models, accuracy):
    print(model, ":", score)
```

```python
import matplotlib.pyplot as plt

plt.bar(
    models,
    accuracy
)

plt.title(
    "Final Model Comparison"
)

plt.ylabel(
    "Accuracy"
)

plt.xticks(rotation=15)

plt.show()
```
