```python
import pandas as pd
import numpy as np
```

```python
races = pd.read_csv("../data/races.csv")
results = pd.read_csv("../data/results.csv")
drivers = pd.read_csv("../data/drivers.csv")
```

```python
results = results[
    ["raceId",
     "driverId",
     "constructorId",
     "grid",
     "positionOrder",
     "points"]
]

results.head()
```

```python
results["winner"] = np.where(
    results["positionOrder"] == 1,
    1,
    0
)

results.head()
```

```python
results["winner"].value_counts()
```

```python
driver_stats = (
    results.groupby("driverId")["points"]
    .mean()
    .reset_index()
)

driver_stats.columns = [
    "driverId",
    "avg_driver_points"
]

driver_stats.head()
```

```python
ml_data = results.merge(
    driver_stats,
    on="driverId",
    how="left"
)

ml_data.head()
```

```python
print(ml_data.shape)
```

# Conclusion

Feature engineering was performed on race results data.

A binary target variable (winner) was created and average driver points were calculated.

The resulting dataset is now suitable for machine learning preparation.
