```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
results = pd.read_csv("../data/results.csv")
```

```python
results = results[
    [
        "raceId",
        "driverId",
        "grid",
        "points",
        "positionOrder"
    ]
]

results["winner"] = np.where(
    results["positionOrder"] == 1,
    1,
    0
)

results.head()
```

```python
driver_stats = (
    results.groupby("driverId")["points"]
    .mean()
    .reset_index()
)

driver_stats.columns = [
    "driverId",
    "avg_driver_points"
]

driver_stats.head()
```

```python
ml_data = results.merge(
    driver_stats,
    on="driverId",
    how="left"
)

ml_data.head()
```

```python
X_old = ml_data[
    [
        "grid",
        "points"
    ]
]

y = ml_data["winner"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_old,
    y,
    test_size=0.2,
    random_state=42
)

old_model = LogisticRegression()

old_model.fit(
    X_train,
    y_train
)

old_predictions = old_model.predict(X_test)

old_accuracy = accuracy_score(
    y_test,
    old_predictions
)

print("Old Model Accuracy:", old_accuracy)
```

```python
X_new = ml_data[
    [
        "grid",
        "points",
        "avg_driver_points"
    ]
]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_new,
    y,
    test_size=0.2,
    random_state=42
)

new_model = LogisticRegression()

new_model.fit(
    X_train,
    y_train
)

new_predictions = new_model.predict(X_test)

new_accuracy = accuracy_score(
    y_test,
    new_predictions
)

print("New Model Accuracy:", new_accuracy)
```

```python
print("Old Model Accuracy :", old_accuracy)
print("New Model Accuracy :", new_accuracy)
```

# Conclusion

A new feature called avg_driver_points was created and added to the model.

Two Logistic Regression models were compared:

1. Using grid and points
2. Using grid, points, and avg_driver_points

This notebook demonstrates how feature engineering can improve machine learning models by providing additional information to the algorithm.
